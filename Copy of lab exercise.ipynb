NUMPY

Declare the Numpy array

```python
import numpy as np
a=np.array([1,2,3])
print(a)
```

Create an array with full of zero values


```python
print(np.zeros(5))
```

Create an array with a Scalar values filled



```python
print(np.full(5, 10))
```

Create an array with random values


```python
print(np.random.rand(5))
```

Reshape and Flattening the array.


```python
a = np.array([[1,2],[3,4]])
print(a.reshape(4,1))
print(a.flatten())
```

Convert an array from one type to another.



```python
a = np.array([1.1, 2.2, 3.3])
print(a.astype('int'))
```

Do slicing operations in an array.


```python
a = np.array([1, 2, 3, 4, 5])
print(a[1:4])
```

Do join functions (join,horizontal join,vertical join and depth join)

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.concatenate([a, b]))
```

Do index retrivel and basic operation with respect to index

```python
a = np.array([1, 2, 3, 4, 5])
print(a[2])
a = np.array([1, 2, 3, 4, 5])
a[2] = 10
print(a)
```

Sorting operation of an array

```python
a = np.array([5, 2, 3, 1, 4])
print(np.sort(a))
```

Filtering operation based on array value


```python
a = np.array([1, 2, 3, 4, 5])
print(a[a > 2])
```

Vector Operation - Addition ,Subtraction, Multiplication and Division


```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

Scalar Operation and Vectorize operation.

```python
a = np.array([1, 2, 3])
print(a + 2)
def myfunc(x):
  return x*2
a = np.array([1, 2, 3])
vfunc = np.vectorize(myfunc)
print(vfunc(a))
```

PANDAS

Declare Empty Dataframe

```python
import pandas as pd
df = pd.DataFrame()
print(df)
```

Declare and print the DataFrame Series

```python
import pandas as pd
s = pd.Series([1, 3, 5, np.nan, 6, 8])
print(s)
```

Add one column and row.

```python
import pandas as pd
df = pd.DataFrame()
df['new_column'] = [1, 2, 3]
print(df)
```

Extract any one column and row based on condition

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 28]}
df = pd.DataFrame(data)
name_column = df['Name']
print(name_column)
```

Do the functions like Sum ,square root ,min,max functiion , sort and merge of values .


```python
import pandas as pd
data = {'Value1': [1, 2, 3], 'Value2': [4, 5, 6]}
df = pd.DataFrame(data)
sum_values = df.sum()
print(sum_values)
```

Create series from array , Dictionary

```python
arr = np.array([1, 2, 3])
s = pd.Series(arr)
print(s)
d = {'a': 1, 'b': 2, 'c': 3}
s = pd.Series(d)
print(s)
```

Create Series using Scalar value,index.

```python
import pandas as pd
s = pd.Series(5, index=[0, 1, 2, 3])
print(s)
```

LOAD AND STORE

Create a dataframe and store the data into specific excel file

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 28]}
df = pd.DataFrame(data)
df.to_excel('my_excel_file.xlsx', index=False)
```

Read and display the excel file data.

```python
import pandas as pd
df = pd.read_excel('my_excel_file.xlsx')
print(df)
```

Display the particular column values , row values and do slicing operations.


```python
import pandas as pd
df = pd.read_excel('my_excel_file.xlsx')
name_column = df['Name']
print(name_column)
```
